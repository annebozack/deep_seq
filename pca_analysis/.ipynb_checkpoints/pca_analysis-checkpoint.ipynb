**Import dependencies**

```python
#import
import numpy as np
import matplotlib.pyplot as plt
from biom import load_table
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA as sklearnPCA
%matplotlib inline
```

**CRISPR spacer PCA**

```python
#load files
file_dir='data/crass/'
file_names=['alien0','alien2','alien7','alien60','bugkiller0','bugkiller2','bugkiller7','bugkiller60']
```

```python
#loading crass dataset into single db
db=[]
for n in np.arange(np.size(file_names)):
    fname=file_dir+file_names[n]+'.csv'
    with open(fname) as f:    
         samp = [x.strip('\n').split(',') for x in f.readlines()]
    db.append(np.zeros(shape=np.shape(samp)))
    db[n]=samp
```

```python
#unpack spacer names and spacer identities
spn=[]
spc=[]
for n in np.arange(np.size(file_names)):
    arrayn=db[n]
    spn_h=[]
    spc_h=[]
    
    #assign spacer identities
    for s in arrayn:
        if s[0] == '""':
            spn_h=np.append(spn_h,s[1].split(':')[1])
    spn.append(np.zeros(shape=np.shape(spn_h)))
    
    
    #assign spacer coverage
    for s in arrayn:
        if s[0] == '""':
            spc_h=np.append(spc_h,int(s[4]))
    spc.append(np.zeros(shape=np.shape(spc_h)))
    spn_h=spn_h[spc_h>1] #retain names with coverage greater than 1 read
    spc_h=spc_h[spc_h>1] #likewise for coverage
    spn[n]=spn_h
    spc[n]=spc_h/np.sum(spc_h)
```

```python
#array with all spacers
sp_all=[]
for a in spn:
    sp_all=np.append(sp_all,a)

#unique spacers
sp_set=list(set(sp_all))
sp_matrix=np.zeros(shape=(np.size(sp_set),8))

#populate spacer matrix
for n in np.arange(np.size(file_names)):
    for sn in np.arange(np.size(sp_set)):
        if sp_set[sn] in spn[n]:
            index=np.where(spn[n] == sp_set[sn])[0][0]
            sp_matrix[sn,n]=spc[n][index]
```

```python
#run PCA
tax_std = StandardScaler().fit_transform(sp_matrix.T)
pca = sklearnPCA(n_components=2)
tra = pca.fit_transform(tax_std)/100

#plotting
col1=['#bdc9e1','#74a9cf','#2b8cbe','#045a8d']
col2=['#fdcc8a','#fc8d59','#e34a33','#b30000']
fig1,ax = plt.subplots(figsize=(6,4))
plt.scatter(tra[0:4,0],tra[0:4,1],c=col1,s=100,edgecolor='',label='alien')
plt.scatter(tra[4:8,0], tra[4:8,1],c=col2,s=100,edgecolor='',label='bugkiller')

plt.xlabel('PC1 ('+"{0:.2f}".format(100*pca.explained_variance_ratio_[0])+'%)')
plt.ylabel('PC2 ('+"{0:.2f}".format(100*pca.explained_variance_ratio_[1])+'%)')
plt.legend(loc=2)
plt.title('PCA: CRISPR spacers')

plt.savefig('crispr_spacers_PCA.pdf', filetype='pdf')
```

**Basic statistics- how many spacers are shared betwen two subjects? ~0.75% when using a bottom threshold**

```python
a_m=np.mean(sp_matrix[:,:4],axis=1)
b_m=np.mean(sp_matrix[:,4:],axis=1)

cnt=0
for i in np.arange(np.size(a_m)):
    if a_m[i] > 0 and b_m[i] > 0:
        cnt=cnt+1
print 'number of shared spacers: '+str(cnt)
print 'total number of unique spacers: '+str(a_m.size)
```
